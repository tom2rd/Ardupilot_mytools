# Ardupilot CopterのFull Parameter ListをCSVにする
html_file_path に　ダウンロード・ファイル名を入れる

```python
from bs4 import BeautifulSoup
# Read the HTML file again
html_file_path = 'Complete Parameter List — Copter documentation.html'
with open(html_file_path, 'r', encoding='utf-8') as f:
    html_content = f.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all elements with the class "toctree-l4" and extract parameter names and links
param_links_alternative = {}
for li in soup.find_all('li', {'class': 'toctree-l4'}):
    anchor = li.find('a')
    if anchor:
        href = anchor['href']
        param_name = anchor.text.strip()
        if param_name:  # Most Ardupilot params are in uppercase
            #param_links_alternative[param_name] = f"https://ardupilot.org/copter/docs/parameters.html{href}"
            param_links_alternative[param_name] = f"{href}"

# Show first 10 items for a quick preview
dict(list(param_links_alternative.items())[:10])
```

```python
import pandas as pd

# キーを2つの部分に分割して新しいリストを作成
data = []
for key, value in param_links_alternative.items():
    param_name, param_description = key.split(': ', 1)
    data.append([param_name, param_description, value])

# リストをDataFrameに変換
df = pd.DataFrame(data, columns=['Parameter Name', 'Parameter Description', 'URL'])

df
```

```python
df.to_csv("all_parameter.csv")
```
