# Ardupilot parameter files convert and Compare side by side
Put the path = "ファイルパス"　*.param


```python
import os
path="ArdupilotCopter4.4parameters"
file_names = os.listdir(path)
file_names
```

```python
import pandas as pd

# Define a function to read a param file and convert it to a dictionary
def read_param_file(file_path):
    param_dict = {}
    with open(file_path, 'r') as f:
        lines = f.readlines()
        for line in lines:
            if line.startswith('#') or line.strip() == '':
                continue
            key, value = line.strip().split(',')
            param_dict[key] = float(value)
    return param_dict

```

```python
all_data = {}
for file_name in file_names:
    all_data[file_name] = read_param_file("./"+path+"/"+file_name)

df = pd.DataFrame(all_data)
df.head()
```
